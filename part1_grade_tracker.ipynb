```python
raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma",       "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair  ",    "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA",       "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ",    "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]
```

```python
cleaned_students = []
```

```python
for student in raw_students:
    # To Clean the name
    name = student["name"].strip().title()
    # To convert roll no into integer
    roll = int(student["roll"])
    # To convert marks_str to list of integers
    marks = list(map(int, student["marks_str"].split(",")))
    #To validate if the name is valid and contains only alphabetic characters
    valid = all(word.isalpha() for word in name.split())
    
if valid:
        print(f"{name}✓ Valid name")
else:
        print(f"{name}✗ Invalid name")

# To store the cleaned/verified data into cleaned_students
cleaned_students.append({
    "name" : name,
    "roll" : roll,
    "marks" : marks
    })

# To print Profile card
print("="*32)
print(f"student : {name}")
print(f"Roll no : {roll}")
print(f"Marks : {marks}")
print("="*32)
```

```python
# To find and print student with roll no 103
for student in cleaned_students:
    print("\nSpecial Output")
    print(student["name"].upper())
    print(student["name"].lower())
```

```python
student_name = "Ayesha Sharma"
subjects = ["Maths","Physics","CS","English","Chemestry"]
marks = [88,72,95,60,78]

print(f"\nMarks Report for {student_name}")

# Calculations like total, highest, Lowest

total = sum(marks)
average = round(total/len(marks), 2)

print(f"\nTotal Marks : {total}")
print(f"Average Marks : {average}")

max_marks = max(marks)
min_marks = min(marks)

max_index = marks.index(max_marks)
min_index = marks.index(min_marks)

print(f"Highest :{subjects[max_index]}({max_marks})")
print(f"Lowest :{subjects[min_index]}({min_marks})")

# For Grades Calculation

def get_grade(score):
    if score >= 90:
        return "A+ Grade"
    elif score >=80:
        return "A Grade"
    elif score >= 70:
        return "B Grade"
    elif score >= 60:
        return "C Grade"
    else:
        return "F Grade"

# To print subject wise marks
for i in range(len(subjects)):
    grade_lable = get_grade(marks[i])
    print(f"{subjects[i]} : {marks[i] ({grade_lable})}")
```

```python
# Task 3 - Class Performaance Summary
```

```python
class_data = [
    ("Ayesha Sharma",  [88, 72, 95, 60, 78]),
    ("Rohit Verma",    [55, 68, 49, 72, 61]),
    ("Priya Nair",     [91, 85, 88, 94, 79]),
    ("Karan Mehta",    [40, 55, 38, 62, 50]),
    ("Sneha Pillai",   [75, 80, 70, 68, 85]),
]

print("\nName    Average   Stattus")
print("-" * 40)

pass_count = 0
fail_count = 0
averages = []

topper_name = ""
topper_avg = 0

for name, marks_list in class_data:
    avg = round(sum(marks_list)/ len(marks_list), 2)
    averages.append(avg)
    
    status = "Pass" if avg>= 60 else "Fail"
    
    if status == "Pass":
        pass_count += 1
    else:
        fail_count +=1
    if avg>topper_avg:
        topper_avg = avg
        topper_name = name
        
    print(f"{name:<18}   {avg: ^7}    {status}")
    
# Final Stats
class_avg = round(sum(averages)/len(averages), 2)

print("\nSummary:")
print(f"Passed: {pass_count}")
print(f"Failed: {fail_count}")
print(f"Topper: {topper_name}({topper_avg})")
print(f"Class Average: {class_avg}")
```

```python
essay = "  python is a versatile language. it supports object oriented, functional, and procedural programming. python is widely used in data science and machine learning.  "

# Step 1
clean_essay = essay.strip()
print("\n1. Clean Essay:\n",clean_essay)

#Step 2
print("\n2. Title case:\n",clean_essay.title())

#Step 3
count = clean_essay.count("python")
print("\n3. 'python' count:",count)

#Step 4
replaced = clean_essay.replace("python", "Python 🐍")
print("\n4. Replaced:\n",replaced)

#Step 5
sentences = clean_essay.split(".")
print("\n5. Sentences List:\n",sentences)

#Step 6
print("\n6. Numbered Sentences:")
for i,sentence in enumerate(sentences,start=1):
    if not sentence.endswith("."):
        sentence += "."
    print(f"{i}.{sentence}")
```
